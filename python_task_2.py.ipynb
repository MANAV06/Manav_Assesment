```python
import pandas as pd

def calculate_distance_matrix(csv_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(csv_file)

    # Create an empty matrix with IDs as columns and index
    unique_ids = sorted(set(df['id_start']) | set(df['id_end']))
    distance_matrix = pd.DataFrame(index=unique_ids, columns=unique_ids)

    # Fill the distance matrix based on the provided distances
    for index, row in df.iterrows():
        id_start, id_end, distance = row['id_start'], row['id_end'], row['distance']

        # Set distance from A to B
        distance_matrix.at[id_start, id_end] = distance

        # Set distance from B to A (bidirectional)
        distance_matrix.at[id_end, id_start] = distance

    # Fill diagonal with zeros (distance from a location to itself)
    for id in unique_ids:
        distance_matrix.at[id, id] = 0

    # Fill in missing distances by calculating A to C using A to B and B to C
    for id_start in unique_ids:
        for id_middle in unique_ids:
            for id_end in unique_ids:
                if pd.isna(distance_matrix.at[id_start, id_end]):
                    if not pd.isna(distance_matrix.at[id_start, id_middle]) and not pd.isna(distance_matrix.at[id_middle, id_end]):
                        distance_matrix.at[id_start, id_end] = distance_matrix.at[id_start, id_middle] + distance_matrix.at[id_middle, id_end]

    return distance_matrix

# Example usage
csv_file_path = 'dataset-3.csv'
result_matrix = calculate_distance_matrix(csv_file_path)
print(result_matrix)
```

```python
import pandas as pd

def unroll_distance_matrix(distance_matrix):
    # Initialize an empty list to store unrolled data
    unrolled_data = []

    # Iterate through the rows and columns of the distance matrix
    for id_start in distance_matrix.index:
        for id_end in distance_matrix.columns:
            # Skip rows where id_start is equal to id_end
            if id_start != id_end:
                # Get the distance value
                distance = distance_matrix.at[id_start, id_end]
                
                # Append the data to the list
                unrolled_data.append({'id_start': id_start, 'id_end': id_end, 'distance': distance})

    # Create a DataFrame from the unrolled data
    unrolled_df = pd.DataFrame(unrolled_data)

    return unrolled_df

# Example usage
# Assuming you already have the distance matrix (result_matrix) from the previous example
result_matrix = calculate_distance_matrix('dataset-3.csv')

# Use the unroll_distance_matrix function
unrolled_df = unroll_distance_matrix(result_matrix)

# Print the resulting DataFrame
print(unrolled_df)
```
